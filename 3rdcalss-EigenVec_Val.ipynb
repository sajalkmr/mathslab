```python
import numpy as np
I=np.array ([[4,3,2],[1,4,1],[3,10,4]])

print("\n Given matrix: \n", I)




eig_val,eig_vec=np.linalg.eig(I)

print("\n Eigen values: \n", eig_val) 
print("\n Eigen vectors: \n", eig_vec)

for i in range(len(eig_val)): 
    print(f"\nEigen vector for eigen value {eig_val[i]} is {eig_vec[i]}\n")
```
