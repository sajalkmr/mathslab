```python
import numpy
matrix1 = numpy.array([[3, -3, 4], [-7, 2, 5], [1, -5, 7]])
print("Matrix")
for row in matrix1:
    print(row)
rank = numpy.linalg.matrix_rank(matrix1)
print("Rank of the given Matrix is : ",rank)
```

```python

i=+1
while True:
  num1=float(input("Enter first number-"))
  num2=float(input("Enter 2nd number-"))
  print("Operations: +,-,*,/")
  selection = input("select operation:")
  if selection =="+":
    print(num1+num2)
  elif selection =="-":
    print(num1-num2)
  elif selection =="*":
    print(num1*num2)
  elif selection =="/":
    print(num1/num2)
  else:
    print("select appropriate operation")


```
