```python
from sympy import plot_implicit, symbols, Eq

x, y = symbols('x y')
```

```python
p1 = plot_implicit(Eq(x**2+y**2,4), (x,-4,4),(y,-4,4), title ='Circle: $x^2+y^2=4$')
```

```python
p2= plot_implicit(Eq(x**(2/3)+y**(2/3), 3**(2/3)), (x,-5,5), (y,-5,5), title="Astroid: x^(2/3)+y^(2/3)=3^(2/3)")
```

```python
p3= plot_implicit(Eq((y**2)*(3-x)-(x**2)*(3+x)),(x,-5,5), (y,-5,5), title="stro")
```

```python
import numpy as np
 
import matplotlib.pyplot as plt 
plt.axes(projection = 'polar')
 
r=3
 
rads = np.arange(0, (2*np.pi), 0.01)
 
#plotting the circle 
for i in rads:
    plt.polar(1, r, "g.")

 

 
plt.show()
```

```python
import numpy as np
from pylab import *
theta=linspace(0,2*np.pi,1000)

x=5*cos(theta)
y=5*sin(theta)
r=sqrt(x**2+y**2)
polar(theta,r,"r")

show()
```

```python
from pylab import *
theta=linspace(0,2*np.pi,1000)
r1=5+5*cos(theta)
polar(theta,r1,"g")
show()
```
