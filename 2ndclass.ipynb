```python
import numpy as np
A=np.matrix([[247,244,464], [564,165,896], [667,667,59]])
B=np.matrix([[54],[64],[54]])
AB=np.concatenate((A,B), axis=1)
rA=np.linalg.matrix_rank(A)
rAB=np.linalg.matrix_rank(AB)
n=A.shape[1]
if (rA==rAB):
  if (rA==n):
    print("The system has unique solution")
    print(np.linalg.solve(A,B))
  else:
    print("the system has infinite solutions")
else:
  print("the sytem of equations is inconsistent")    
```

```python
def calculate_prime_factors(N):
    prime_factors = set()
    if N % 2 == 0:
        prime_factors.add(2)
    while N % 2 == 0:
        N = N // 2
        if N == 1:
            return prime_factors
    for factor in range(3, N + 1, 2):
        if N % factor == 0:
            prime_factors.add(factor)
            while N % factor == 0:
                N = N // factor
                if N == 1:
                    return prime_factors


input_number = 200989898989898989898989898989899999999999999999999999999

output = calculate_prime_factors(input_number)
print("Prime factors of {} are {}".format(input_number, output))
```
